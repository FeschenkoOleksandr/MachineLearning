```python
#Завдання_1 Вивести перших 5 рядків.
import pandas as pd

file_path = '/content/titanic (2).csv'
df = pd.read_csv(file_path)
df.head()
```

```python
#Завдання_2 Визначити розмір датасета
df.shape
```

```python
#Завдання_3 Визначити тип даних
print("\n Типи даних кожного стовпця:\n", df.dtypes)
```

```python
#Завдання_4 Визначити наявність пропущених значень. При наявності, замінити пропущені значення на середнє значення
df.isnull().sum()

```

```python
#Завдання_4 Визначити наявність пропущених значень. При наявності, замінити пропущені значення на середнє значення
df['Age'].mean(skipna=True)
df["Age"].fillna(df["Age"].median(skipna=True), inplace=True)
df["Fare"].fillna(df['Fare'].value_counts().idxmax(), inplace=True)
```

```python
#Завдання_8 Видалити стовпчик Cabin
df.drop('Cabin', axis=1, inplace=True)
```

```python
#Завдання_5  Ще раз перевірити наявність пропущених значень.
df.isnull().sum()
```

```python
df.head()
```

```python
#Завдання_9 Сформувати датасет з обраними стовпцями
df = df[['Survived', 'Pclass', 'Sex', 'Age', 'Fare']]
df.head()
```

```python
#Завдання_7 Вивести описову статистику датасету describe()
df.describe()
```

```python
#Завдання_10 Замінити бінарні ознаки (Стать) на 0 і 1.
df['Sex'] = df['Sex'].replace({'female': 1, 'male' : 0}).astype(int)
df.dtypes
```

```python
df.head(10)
```

```python
#14.Аналіз виживання залежно від статі: Обчисліть відсоток виживання для кожної статі.
survival_by_sex = df.groupby('Sex')['Survived'].mean() * 100
print(survival_by_sex)
```

```python
#15.Обчисліть відсоток виживання для кожного класу (Pclass).
survival_by_class = df.groupby('Pclass')['Survived'].mean() * 100
print(survival_by_class)
```

```python
#16. Визначте середній вік тих, хто вижив, і тих, хто не вижив.
mean_age_survived = df[df['Survived'] == 1]['Age'].mean()
mean_age_not_survived = df[df['Survived'] == 0]['Age'].mean()
print(f"Середній вік тих, хто вижив: {mean_age_survived}")
print(f"Середній вік тих, хто не вижив: {mean_age_not_survived}")
```

```python
#17. Розподіліть пасажирів на групи за рівнями тарифів (Fare) і обчисліть рівень виживання для кожної групи.
df['FareGroup'] = pd.qcut(df['Fare'], q=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])


survival_by_fare_group = df.groupby('FareGroup', observed=False)['Survived'].mean() * 100
print(survival_by_fare_group)

```

```python
#18. Аналіз класу та тарифу: Визначте середній тариф (Fare) для кожного
mean_fare_by_class = df.groupby('Pclass')['Fare'].mean()
print(mean_fare_by_class)
```

```python
#19. Обчисліть середній вік пасажирів для кожного класу (Pclass).

mean_age_by_class = df.groupby('Pclass')['Age'].mean()
print(mean_age_by_class)
```

```python
import matplotlib.pyplot as plt

#20.Побудуйте гістограму розподілу віку для тих, хто вижив, і тих, хто не вижив.
plt.figure(figsize=(12, 6))
plt.hist(df[df['Survived'] == 1]['Age'].dropna(), bins=20, alpha=0.5, label='Вижили')
plt.hist(df[df['Survived'] == 0]['Age'].dropna(), bins=20, alpha=0.5, label='Не вижили')

plt.title('Розподіл віку для тих, хто вижив і тих, хто не вижив')
plt.xlabel('Вік')
plt.ylabel('Кількість пасажирів')
plt.legend()
plt.show()
```

```python
#21.Обчисліть відсоток виживання для кожної комбінації статі та класу (наприклад, жінки в 1-му класі, чоловіки в 3-му класі).
survival_by_sex_class = df.groupby(['Sex', 'Pclass'])['Survived'].mean() * 100
print(survival_by_sex_class)
```

```python
#22.Обчисліть кореляцію між усіма числовими змінними (Survived, Pclass, Age, Fare). Які змінні найбільш сильно корелюють з виживанням ?
correlation_matrix = df[['Survived', 'Pclass', 'Age', 'Fare']].corr()
print(correlation_matrix)
```
