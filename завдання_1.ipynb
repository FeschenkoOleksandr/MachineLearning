```python
import numpy as np

# Завдання 1 Створіть одномірний масив (вектор) зі значеннями квадратних коренів з кожного натурального числа, починаючи з 1, та виведіть його значення.
vector_1 = np.sqrt(np.arange(1, 11))  # Квадратні корені натуральних чисел від 1 до 10
print("Завдання 1: вектор квадратних коренів натуральних чисел від 1 до 10")
print(vector_1)
```

```python
# Завдання 2 Створіть двовимірний масив (матрицю) розміром 3x3, заповніть його значеннями коренів 4 ступеня перших дійсних чисел.
matrix_2 = np.power(np.arange(1, 10).reshape(3, 3), 1/4)  # Корені 4 ступеня перших дійсних чисел
print("\nЗавдання 2: матриця 3x3 зі значеннями коренів 4 ступеня перших дійсних чисел")
print(matrix_2)
```

```python
# Завдання 3 Створіть масив розміром 3x3, заповніть його випадковими натуральними числами від 1 до 50 та знайдіть суму елементів кожного рядка.
array_3 = np.random.randint(1, 51, (3, 3))  # Випадкові натуральні числа від 1 до 50
суми_рядків = array_3.sum(axis=1)
print("\nЗавдання 3: array 3x3 з випадковими натуральними числами від 1 до 50")
print(array_3)
print("суми елементів кожного рядка:", суми_рядків)
```

```python
# Завдання 4 Створіть масив розміром 4x4, заповніть його випадковими дійсними числами від -0.9 до 0.9 та виведіть його значення.
array_4 = np.random.uniform(-0.9, 0.9, (4, 4))  # Випадкові дійсні числа від -0.9 до 0.9
print("\nЗавдання 4: масив 4x4 з випадковими дійсними числами від -0.9 до 0.9")
print(array_4)
```

```python
# Завдання 5 Створіть два одномірні масиви розміром 7, заповніть їх випадковими цілими числами в діапазоні від 1 до 5 та виконайте на них поелементні операції додавання, віднімання та множення.
array_5_1 = np.random.randint(1, 6, 7)  # Перший array
array_5_2 = np.random.randint(1, 6, 7)  # Другий array

додавання = array_5_1 + array_5_2
віднімання = array_5_1 - array_5_2
множення = array_5_1 * array_5_2

print("\nЗавдання 5: Операції над arrayами")
print("Перший масив:", array_5_1)
print("Другий масив:", array_5_2)
print("Додавання:", додавання)
print("Віднімання:", віднімання)
print("Множення:", множення)
```

```python
# Завдання 6 Створіть дві матриці розміром 3x3, заповніть їх довільними цілими числами від 0 до 5 та знайдіть їх добуток.
matrix_6_1 = np.random.randint(0, 6, (3, 3))  # Перша matrix
matrix_6_2 = np.random.randint(0, 6, (3, 3))  # Друга matrix

добуток_6 = np.dot(matrix_6_1, matrix_6_2)

print("\nЗавдання 6: Добуток двох матриць 3x3")
print("Перша матриця:")
print(matrix_6_1)
print("Друга матриця:")
print(matrix_6_2)
print("Добуток:")
print(добуток_6)
```

```python
# Завдання 7 Створіть дві матриці розміром 2x3 та 3x4, заповніть їх випадковими цілими числами в діапазоні від 1 до 7 та перемножте їх між собою.
matrix_7_1 = np.random.randint(1, 8, (2, 3))  # Перша matrix 2x3
matrix_7_2 = np.random.randint(1, 8, (3, 4))  # Друга matrix 3x4

добуток_7 = np.dot(matrix_7_1, matrix_7_2)

print("\nЗавдання 7: Добуток матриць 2x3 та 3x4")
print("Перша матриця:")
print(matrix_7_1)
print("Друга матриця:")
print(matrix_7_2)
print("Добуток:")
print(добуток_7)
```

```python
# Завдання 8 Створіть матрицю розміром 2x2, заповніть її випадковими цілими числами в діапазоні від 1 до 15 та знайдіть її обернену матрицю.

matrix_8 = np.random.randint(1, 16, (2, 2))  # matrix 2x2

inverse_matrix_8 = np.linalg.inv(matrix_8)

print("\nЗавдання 8: Обернена матриця 2x2")
print("matrix:")
print(matrix_8)
print("Обернена матриця:")
print(inverse_matrix_8)
```

```python
# Завдання 9 Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами в діапазоні від 3 до 15 та транспонуйте її.
matrix_9 = np.random.randint(3, 16, (3, 3))  # matrix 3x3


transposed_matrix_9 = np.transpose(matrix_9)

print("\nЗавдання 9: Транспонована матриця 3x3")
print("матриця:")
print(matrix_9)
print("Транспонована матриця:")
print(transposed_matrix_9)
```

```python
# Завдання 10 Створіть матрицю розміром 3x4 та вектор розміром 4x1, заповніть матрицю випадковими цілими числами в діапазоні від 1 до 10, а вектор - випадковими цілими числами в діапазоні від 1 до 10, і знайдіть добуток матриці на вектор.
matrix_10 = np.random.randint(1, 11, (3, 4))  # matrix 3x4
vector_10 = np.random.randint(1, 11, (4, 1))  # vector 4x1

добуток_10 = np.dot(matrix_10, vector_10)

print("\nЗавдання 10: Добуток матриці 3x4 на вектор 4x1")
print("матриця:")
print(matrix_10)
print("вектор:")
print(vector_10)
print("Добуток:")
print(добуток_10)
```

```python
# Завдання 11 Створіть дві матриці розміром 2x3 та 3x3, заповніть першу матрицю випадковими цілими числами в діапазоні від 1 до 5, а другу - випадковими дійсними числами в діапазоні від -1 до 1 та знайдіть добуток двох елементів першої матриці.
matrix_11_1 = np.random.randint(1, 6, (2, 3))  # matrix 2x3
matrix_11_2 = np.random.uniform(-1, 1, (3, 3))  # matrix 3x3

добуток_елементів_11 = matrix_11_1[0, 0] * matrix_11_1[1, 1]

print("\nЗавдання 11: Добуток двох елементів першої матриці 2x3")
print("Перша матриця:")
print(matrix_11_1)
print("Друга матриця:")
print(matrix_11_2)
print("Добуток елементів:", добуток_елементів_11)
```

```python
# Завдання 12 Створіть матрицю розміром 3x4 та заповніть її випадковими дійсними числами в діапазоні від -0.9 до 0.9 та обчисліть її визначник.
matrix_12 = np.random.uniform(-0.9, 0.9, (3, 4))  # matrix 3x4

# Визначник можна обчислити лише для квадратної матриці, тому тут ми будемо обчислювати псевдо-визначник (через SVD)
U, S, Vt = np.linalg.svd(matrix_12)
псевдо_визначник_12 = np.prod(S)

print("\nЗавдання 12: Псевдо-визначник матриці 3x4")
print("matrix:")
print(matrix_12)
print("Псевдо-визначник:", псевдо_визначник_12)
```

```python
# Завдання 13 Створіть масив розміром 2x3 та вектор розміром 3x1, заповніть масив випадковими цілими числами в діапазоні від 1 до 5, а вектор - випадковими дійсними числами в діапазоні від -1 до 1, та знайдіть їх добуток.
array_13 = np.random.randint(1, 6, (2, 3))  # array 2x3
vector_13 = np.random.uniform(-1, 1, (3, 1))  # vector 3x1

добуток_13 = np.dot(array_13, vector_13)

print("\nЗавдання 13: Добуток масиву 2x3 на вектор 3x1")
print("array:")
print(array_13)
print("vector:")
print(vector_13)
print("Добуток:")
print(добуток_13)
```

```python
# Завдання 14 Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами в діапазоні від 1 до 10 та знайдіть добуток її головної діагоналі.
matrix_14 = np.random.randint(1, 11, (3, 3))  # matrix 3x3

добуток_діагоналі_14 = np.prod(np.diag(matrix_14))

print("\nЗавдання 14: Добуток головної діагоналі матриці 3x3")
print("matrix:")
print(matrix_14)
print("Добуток головної діагоналі:", добуток_діагоналі_14)
```

```python
# Завдання 15 Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами в діапазоні від 1 до 5 та знайдіть вектор-стовпчик, що містить суму елементів кожного стовпця матриці.
matrix_15 = np.random.randint(1, 6, (3, 3))  # matrix 3x3

vector_sum_15 = matrix_15.sum(axis=0).reshape(-1, 1)

print("\nЗавдання 15: вектор-стовпчик сум елементів кожного стовпця матриці 3x3")
print("matrix:")
print(matrix_15)
print("вектор-стовпчик сум елементів кожного стовпця:")
print(vector_sum_15)
```
